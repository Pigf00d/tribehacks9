```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.action_chains import ActionChains
from time import sleep
import pandas as pd
```

```python
arrayOfTeachers = []
options = Options()
options.headless = True
options.add_argument("--window-size=1920,1200")
```

```python
dfOfTeachers = pd.DataFrame(columns=['Name', 'Quality', 'Difficulty', 'Would Take Again', 'ID'])
DRIVER_PATH = "C:\\Users\\theso\\chromedriver-win64"
driver = webdriver.Chrome(options=options)
driver.get('https://www.ratemyprofessors.com/search/professors/269?q=*')
while len(arrayOfTeachers) < 10:
    print("new page")
    sleep(2)
    h2 = driver.find_elements(By.CLASS_NAME, 'TeacherCard__StyledTeacherCard-syjs0d-0')
    for i in h2:
        name = i.text.split('\n')[3]
        quality = i.text.split('\n')[1]
        difficulty = i.text.split('\n')[8]
        wouldTakeAgain = i.text.split('\n')[5]
        #teacherID = i.get_attribute('data-id')
        if name not in dfOfTeachers['Name'].values: #ig this isn't working
            print(dfOfTeachers['Name'].values)
            dfOfTeachers = pd.concat([dfOfTeachers, pd.DataFrame({'Name': [name], 'Quality': [quality], 'Difficulty': [difficulty], 'Would Take Again': [wouldTakeAgain]})], ignore_index=True)
    #see if I can click the button
    button = None
    try:
        button = driver.find_element(By.CLASS_NAME, 'Buttons__Button-sc-19xdot-1')
    except:
        print("Button not found")
    if button is not None:
        ActionChains(driver).move_to_element(button).perform()  # Move to the button
        button.click()
    sleep(2) 
    sleep(2)
driver.quit()
```
